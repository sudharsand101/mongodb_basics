```python
#)1. importing mongodb module
import pymongo
```

```python
#)2. creating connection between jupyter notebook to mongodb
connection = pymongo.MongoClient('mongodb://127.0.0.1:27017/')
```

```python
#)3. table(collection) creation in mongodb database
db = connection['D101']
collection = db['course']
```

```python
#) 4. inserting one doc
data = {
    'name':'madhur',
    'course':'DS',
    'batch':'2A',
    'institute':'ABC'
}
db.collection.insert_one(data)
```

```python
#) importing csv data as dataframe
import pandas as pd
df = pd.read_csv(r"D:\Sudharsan\Guvi_Data science\DS101_Sudharsan\Mainboot camp\data\diabetes.csv")
df
```

```python
#) creating diabetes collection
db = connection['D101']
col = db['diabetes']

#) 8. inserting data from csv file to mongodb
my_data = []

for index in df.index:
    row = df.loc[index]
    row = row.to_dict()
    my_data.append(row)

col.insert_many(my_data)
```

```python
#) 9. to show inserted data from csv
for doc in col.find():
    print(doc)
```

```python
#)10. not displaying id columns
for doc in col.find({},{'_id':0}):
    print(doc)
```

```python
#) 11. removing one column data
for doc in col.find({},{'_id':0,'SkinThickness':0}):
    print(doc)
```

```python
#) 12.removing 2 columns
for doc in col.find({},{'_id':0,'SkinThickness':0,'BMI':0}):
    print(doc)
```

```python
#) 13. displaying only 2 columns data
for doc in col.find({},{'_id':0,'Pregnancies':1,'Glucose':1}):
    print(doc)
```

```python
#) 14.displaying particular row and particular coloumn data
for doc in col.find({'Pregnancies': 1.0,'Insulin': 0.0},{'Pregnancies': 1,'Insulin':1,'Age':1,'_id':0}):
    print(doc)
```

```python
#) 15.displaying particular 2 column data using $and
condition = {'$and':[{'BloodPressure': 70.0},{'BMI': 28.9}]}
for doc in col.find(condition,{'_id':0}):
    print(doc)
```

```python
#) 16. displaying two particular data from one column using $or
condition = {'$or':[{'Glucose': 102.0},{'Glucose': 141.0}]}
for doc in col.find(condition,{'_id':0}):
    print(doc)
```

```python
#) 17. displaying particular data using $gt(greater than)
condition = {'BloodPressure':{'$gt':74},'Pregnancies': 0.0,'SkinThickness': 0.0}
for doc in col.find(condition,{'_id':0,}):
    print(doc)
```

```python
#) 18. displaying particular data using $gt(greater than or equal to)
condition = {'BloodPressure':{'$gte':76},'Pregnancies': 0.0,'SkinThickness': 0.0}
for doc in col.find(condition,{'_id':0,}):
    print(doc)
```

```python
#) 19. displaying particular data using $lt(less than)
condition = {'BloodPressure':{'$lt':76},'Pregnancies': 0.0,'SkinThickness': 0.0}
for doc in col.find(condition,{'_id':0,}):
    print(doc)
```

```python
#) 20. displaying particular data using $lte(less than or equal to)
condition = {'BloodPressure':{'$lte':76},'Pregnancies': 0.0,'SkinThickness': 0.0}
for doc in col.find(condition,{'_id':0,}):
    print(doc)
```

```python
#) 21. displaying particular data using $ne(not equal to)
condition = {'BloodPressure':{'$ne':76},'Pregnancies': 0.0,'SkinThickness': 0.0}
for doc in col.find(condition,{'_id':0,}):
    print(doc)
```

```python
#) 22. displaying particular data using two conditions $or,$ne(not equal to)
condition = {'$or':[{'BloodPressure':{'$ne':76}},{'Age': 21.0}],'SkinThickness': 0.0}
for doc in col.find(condition,{'_id':0}):
    print(doc)
```

```python
#) 23. replacing one row data with new value in one column
old = {'SkinThickness': 0.0}
new = {'$set':{'SkinThickness': 1.0}}

col.update_one(old,new)
for doc in col.find({},{'_id':0,'SkinThickness': 1}):
    print(doc)
```

```python
#) 24. replacing many row data with new value
old = {'SkinThickness': 0.0}
new = {'$set':{'SkinThickness': 1.0}}

col.update_many(old,new)
for doc in col.find({},{'_id':0,'SkinThickness': 1}):
    print(doc)
```

```python
#) 25. not displaying particular row data from two columns
old = {'Pregnancies': 0.0}
new = {'$unset':{ 'SkinThickness': 1.0}}

col.update_many(old,new)
for doc in col.find({},{'_id':0}):
    print(doc)
```

```python
#) 26. adding new column for particular row data
old = {'Pregnancies': 0.0}
new = {'$set':{ 'generation': "GenZ"}}

col.update_many(old,new)
for doc in col.find({},{'_id':0}):
    print(doc)
```

```python
#) 27. deleting one doc
condition ={'Pregnancies': 6.0}
col.delete_one(condition)

for doc in col.find({},{'_id':0}):
    print(doc)
```

```python
#) 28. deleting many doc
condition ={'Pregnancies': 6.0}
col.delete_one(condition)

for doc in col.find({},{'_id':0}):
    print(doc)
```
